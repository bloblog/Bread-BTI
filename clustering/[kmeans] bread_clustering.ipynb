```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

## 빵 클러스터링

```python
bread = pd.read_excel("빵 특성 수치화(10.27).xlsx")
x = bread.iloc[:, 1:] 
x.head()
```

### 1. K-means로 클러스터링

```python
km = KMeans(n_clusters=5) # 5개의 종류로 클러스터링 해라
km.fit(x)

cluster = km.predict(x)
cluster
```

### 2. PCA(차원축소)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # 2차원으로 축소 
pca.fit(x)
pc = pca.transform(x)

import seaborn 
seaborn.scatterplot(x=pc[:, 0], y=pc[:, 1], hue=cluster,
                    palette=['#ff7f0e', '#2ca02c', '#6fe1ed', '#ca63f2', '#424242'])
```
