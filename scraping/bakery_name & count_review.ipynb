**모듈 임포트**

```python
from bs4 import BeautifulSoup

import selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import time
import pandas as pd

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys 
from selenium.common.exceptions import NoSuchElementException 
```

**크롬 드라이버 설치**

```python
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
```

```python
# 서울 00 맛집 검색
driver.get("https://pcmap.place.naver.com/restaurant/list?query=%EC%84%9C%EC%9A%B8%20%EC%8B%9D%EB%B9%B5%20%EB%A7%9B%EC%A7%91&x=126.957802153937&y=37.477136000000286&bounds=126.9540685189902%3B37.473329982569595%3B126.97011885773594%3B37.48077154481187&entry=pll&rank=%EB%A6%AC%EB%B7%B0%EB%A7%8E%EC%9D%80&clientX=126.9424278&clientY=37.4853086&deviceType=pcmap&lastSelected=%EC%A0%95%EB%A0%AC%EC%98%B5%EC%85%98")

bakery_list = []
reviews = []
    
for i in range(3):
    scroll_area=driver.find_element(By.XPATH, '//*[@id="_pcmap_list_scroll_container"]')
    action = webdriver.ActionChains(driver)
    action.move_to_element(scroll_area).perform()

    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll_area)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll_area)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll_area)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll_area)
    driver.execute_script("arguments[0].scrollTop = arguments[0].scrollHeight", scroll_area)
    time.sleep(1)

    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")

    bakery_select = soup.select("span.TYaxT")
    for i in range(len(bakery_select)):
        bakery_list.append(bakery_select[i].text)


    review_select = soup.select("span.h69bs")
    for i in range(len(review_select)):
        reviews.append(review_select[i].text)

    driver.find_element(By.CSS_SELECTOR, '#app-root > div > div.XUrfU > div.zRM9F > a:nth-child(7) > svg').click()
print(bakery_list)
```

```python
review_count = [s for s in reviews if '방문자리뷰' in s]
print(review_count)
```

```python
len(bakery_list) == len(review_count)
# True -> 진행
```

```python
data = []
for m,n in zip(bakery_list, review_count):
    data.append([m,n])

df = pd.DataFrame(data, columns=["검색어", "방문자리뷰수"])
df.to_excel('식빵빵집.xlsx')
```
