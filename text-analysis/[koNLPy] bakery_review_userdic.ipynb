### 빵 그룹별로 데이터프레임 통합
- 각자 불러와서
- `concat` 으로 합치고
- 데이터프레임 형태 수정
- NaN 값 처리 한 번 더

```python
df1 = pd.read_excel('에그타르트_전처리완.xlsx')
df2 = pd.read_excel('수플레_전처리완.xlsx')
df3 = pd.read_excel('카스테라_전처리완.xlsx')
```

```python
df = pd.concat([df1,df2,df3],axis=1)
df = df.drop(columns = ['Unnamed: 0'])
df = df.fillna("")
```

```python
df = df.transpose()
cols = list(range(200))
df['review'] =df[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

df = df.drop(df.columns[:-1], axis=1)

df
```

### konlpy 로 데이터 전처리
- 어간추출
- 조사 어미 구두점 제외
- `review` 컬럼에 다시 넣기

```python
import numpy as np
import konlpy

for i in range(len(df)):
    okt = konlpy.tag.Okt()
    clean_words = []
    for word in okt.pos(df['review'][i], stem=True):
        if word[1] not in ['Josa', 'Eomi', 'Punctuation']:
            clean_words.append(word[0])
    df['review'][i] = ' '.join(clean_words)
df.head()
```

### 전처리한 형태소 분석 결과 데이터프레임화
- `df.result` 데이터프레임 만들어서 분석 결과 넣는다.

```python
name = []
result = []
for i in range(len(df)):
    tokens_ko = okt.morphs(df['review'][i], stem = True)
    ko = nltk.Text(tokens_ko)
    
    name.append(df.index[i])
    result.append(ko.vocab().most_common())
    
df_result = pd.DataFrame(zip(name, result), columns=['빵집','빈도'])
df_result
```

```python
df_result.to_excel('3조- 에그타르트, 수플레, 카스테라.xlsx')
```
