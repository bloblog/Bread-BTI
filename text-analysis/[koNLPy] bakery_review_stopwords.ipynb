### 필요한 패키지 및 모듈 다운

```python
import pandas as pd
import numpy as np

import konlpy
from konlpy.tag import Okt
okt = Okt()

import nltk
nltk.download('punkt')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
```

### 엑셀파일 가져오기
- 엑셀 파일 통합
- 데이터프레임 정리
    - 컬럼삭제
    - NaN값 공백 대체
    - 행열 전환
    - 셀 통합
    - 리뷰 셀만 남기기

```python
df1 = pd.read_excel('스콘_전처리완.xlsx')
df2 = pd.read_excel('프레첼_전처리완.xlsx')

df = pd.concat([df1,df2],axis=1)
df = df.drop(columns = ['Unnamed: 0'])
df = df.fillna("")

df = df.transpose()
cols = list(range(200))
df['review'] =df[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)

df = df.drop(df.columns[:-1], axis=1)

df
```

### 어간추출, 조사 어미 구두점 제거

```python
for i in range(len(df)):
    okt = konlpy.tag.Okt()
    clean_words = []
    for word in okt.pos(df['review'][i], stem=True):
        if word[1] not in ['Josa', 'Eomi', 'Punctuation']:
            clean_words.append(word[0])
    df['review'][i] = ' '.join(clean_words)
df.head()
```

### 불용어 적용하기

```python
stop_words=불용어.values

for i in range(len(df)): 
    result =[]
    text = df['review'][i]
    word_tokens = word_tokenize(text)
    for w in word_tokens: 
        if w not in stop_words: 
            result.append(w)
    df['review'][i] = result.copy()
df
```

### 전처리한 자료 빈도 계산

```python
name = []
result = []
for i in range(len(df)):
    tokens_ko = okt.morphs(str(df['review'][i]))
    ko = nltk.Text(tokens_ko)
    
    name.append(df.index[i])
    result.append(ko.vocab().most_common())
    
df_result = pd.DataFrame(zip(name, result))
df_result.to_excel('불용어1차제거_8조.xlsx')
```
